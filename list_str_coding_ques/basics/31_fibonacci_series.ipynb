## Fibonacci series

```python
# First way - Assuming that temp list already have these two values - 0, 1
n=10
temp = [0,1]
for i in range(n):
    val = temp[-1] + temp[-2]
    temp.append(val)
print(temp)
```

```python
# Second way - if temp list is empty then: 
temp=[]
n=10

for i in range(n):
    if i==0 or i==1:
        temp.append(i)
    else:
        val = temp[-1] + temp[-2]
        temp.append(val)
print(temp)
```

```python
# third way - using variable instead of list - more optimized
a, b = 0,1
for i in range(10):
    print(a, end=' ')
    a, b = b, a+b
```

```python
# forth way - using generator
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

print(list(fib(10)))
```

```python
# firth way - using while loop
num=150
a,b=0,1
print(a, end=' ')
while b<num:
    print(b, end=' ')
    a,b = b, a+b  # swapping without using 3rd variable

```

```python
# sixth way by using recursion - interviewer prefer this

def recur_fibo(n: int):
    if n<=1:
        return n
    else:
        return (recur_fibo(n-1) + recur_fibo(n-2))
n=10
for i in range(n):
    print(recur_fibo(i), end=' ')
```
