Two questions -
1. Check if a string is palindrome or not
2. Check if a number if palindrome or not

### String

```python
# first way

name = 'Ashish'

def is_palindrome1(name: str):
    return name[::-1] == name

print(is_palindrome1(name))
```

```python
# second way - by using new variable/for loop

def is_palindrome2(name: str):
    
    # manually reversing string
    new_str = ''
    for i in name:
        new_str = i+new_str
    return new_str == name

name = 'yty'

print(is_palindrome2(name))
```

```python
# Third way - by using indexes/for loop

def is_palindrome3(name: str):
    
    word_len = len(name)
    for i in range(word_len):
        if name[i] != name[word_len-1-i]:
            return False
    return True

name = 'yty'

print(is_palindrome3(name))
```

```python
# forth way - by using in built methods - reverse/join

def is_palindrome4(name: str):
    
    lst_name = name.split()
    lst_name.reverse()  # reverse doesnt return anything
    return ''.join(lst_name) == name


name = 'yty'

print(is_palindrome4(name))
```

```python
# fifth way - by using in built methods - reversed/join

def is_palindrome5(name: str):
    temp = reversed(name)  # reversed function gives us an iterable object.
    return ''.join(temp) == name


name = 'yty'

print(is_palindrome5(name))
```

```python
# sixth way - by using while loop

def is_palindrome6(name: str) -> bool:
    first_idx = 0
    last_idx = len(name) - 1

    while first_idx<last_idx:
        if name[first_idx] != name[last_idx]:
            return False
        first_idx+=1
        last_idx-=1
    return True



name = 'yty'

print(is_palindrome6(name))
```

### Number

```python
# first way
# interviewer dont like this sol, but here it is by converting to str and using slicing
def check_palindrome(n: int):

    return str(n)[::-1] == str(n)

check_palindrome(555)
```

```python
# using while loop - best approach for numbers
def check_palindrome(n: int):
    temp = n
    rev=0
    while temp>0:
        unit_digit = temp%10
        rev = rev*10 + unit_digit
        temp=temp//10
    if rev==n:
        return True
    else:
        return False

check_palindrome(555)
```
